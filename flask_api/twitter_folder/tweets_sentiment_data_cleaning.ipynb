```
def cleanText( text):
  
  text = re.sub(r'@[A-Za-z0-9]+', '', text) # remove @mentions
  text = re.sub(r'#', '', text) # remove # sign
  text = re.sub(r'RT[\s]+', '', text) # remove RT
  text = re.sub(r'http\S+', '', text)  # remove hyperlinks
  text = text.lower()   # convert to lower case
  for punct in string.punctuation:  # remove punctuations
    text = text.replace(punct, " ")

  # for 's and 'm, 're , 
  for stray in [" s ", " re ", " m "]:
    text = text.replace(stray, " ")

  text = text.strip()   # remove whitespace
  return text
  
print(string.punctuation)
```

```
cleaned_df = textAndLabels_df.copy()
# applying data cleaning
cleaned_df['text'] = textAndLabels_df['text'].apply(cleanText)

# show cleaned text
printTweets(cleaned_df, 1000)
```

Lemmatizing (done and saved as csv, so commented out)

```
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```

lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("cars"))
```

```
i =0
def lemma_function(text):
  global i
  dummy = []
  for word in text.split(" "):
    if len(word) >0:
      dummy.append(lemmatizer.lemmatize(word))
  print(i, dummy)
  i += 1
  return ' '.join(dummy)
```

```
cleaned_df['lemmatized'] = cleaned_df['text'].apply(lambda x: lemma_function(x))

```
